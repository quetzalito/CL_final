```python
import pandas as pd

from sklearn.metrics import classification_report, confusion_matrix
```

```python
import os
os.chdir('/Users/Zila/opt/anaconda3/lib/python3.9/site-packages/py_tat_morphan')
```

```python
art_df = pd.read_csv('articles_clean.csv')
art_df.head()
```

```python
art_df.drop(['Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0', 'link', 'title', 'text'], axis =1, inplace=True) 
art_df.head()
```

```python
#векторизация
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(art_df['lemmas'])
```

```python
# разделение на обучающую и тестовую выборку
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, art_df["category"], test_size=0.15, random_state=25)
```

```python
print(X_train.shape, X_test.shape)
```

```python
DECISION TREE
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
```

```python
tree_y_pred = tree_clf.predict(X_test)
```

```python
len(tree_y_pred)
```

```python
print(classification_report(y_test, tree_y_pred))
```

```python
print(confusion_matrix(y_test, tree_y_pred))
```

```python
BAGGING
```

```python
from sklearn.ensemble import RandomForestClassifier 
forest = RandomForestClassifier(n_estimators=1000, verbose=1)
```

```python
forest.fit(X_train, y_train)
```

```python
forest_y_pred = forest.predict(X_test)
```

```python
len(forest_y_pred)
```

```python
#evaluation
print(classification_report(y_test, forest_y_pred))
```

```python
confusion_matrix(y_test, forest_y_pred)
```

```python
BOOSTING
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=1000, random_state=10, verbose=10)
```

```python
gb_clf.fit(X_train, y_train)
```

```python
gb_y_pred = gb_clf.predict(X_test)
```

```python
print(classification_report(y_test, gb_y_pred))
```

```python
print(confusion_matrix(y_test, gb_y_pred))
```
